```python
data = {
    'City': ['New York City', 'Washington, D.C.', 'San Francisco', 'Berlin', 'Los Angeles'],
    'Region': ['New York', 'District of Columbia', 'California', '', 'California'],
    'Country': ['United States of America', 'United States of America', 'United States of America', 'Germany', 'United States of America'],
    'AirQuality': [46.816038, 66.129032, 60.514019, 62.364130, 36.621622],
    'WaterPollution': [49.504950, 49.107143, 43.000000, 28.612717, 61.299435]
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.barh(df['City'], df['AirQuality'], color='skyblue')
plt.xlabel('Air Quality')
plt.ylabel('City')
plt.title('Distribution of Air Quality Across Cities')
plt.grid(axis='x')
plt.show()
```

```python
data = {
    'City': ['New York City', 'Washington, D.C.', 'San Francisco', 'Berlin', 'Los Angeles'],
    'Region': ['New York', 'District of Columbia', 'California', '', 'California'],
    'Country': ['United States of America', 'United States of America', 'United States of America', 'Germany', 'United States of America'],
    'AirQuality': [46.816038, 66.129032, 60.514019, 62.364130, 36.621622],
    'WaterPollution': [49.504950, 49.107143, 43.000000, 28.612717, 61.299435]
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.bar(df['Region'], df['WaterPollution'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Water Pollution')
plt.title('Water Pollution Across Countries')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.show()
```
